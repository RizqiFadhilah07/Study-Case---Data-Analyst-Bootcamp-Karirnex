## 🎯 Objective
Kita akan menganalisis dataset penjualan FMCG nyata dari perusahaan yang menjual produk kebutuhan sehari-hari melalui berbagai channel (Offline, Online Toko Oren, Online Toko Hijau).

1. Berapa total revenue dan komposisi per kategori produk?

3. Bagaimana tren penjualan per bulan?
4. Produk apa yang paling laris?
5. Kota mana dengan penjualan tertinggi?
6. Bagaimana distribusi status order dan rating pelanggan?
7. Top 10 Pelanggan Paling Aktif
8. Kategori Terlaris pada Setiap Channel

### Import Datasets and Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load dataset
# Jika di Google Colab: upload file atau mount Google Drive terlebih dahulu
df_raw = pd.read_csv("Data FMCG_BT Data Analyst - Messy.csv", sep=';')
print(f'Shape awal : {df_raw.shape}')
print(f'Kolom      : {df_raw.columns.tolist()}')
print()
df_raw.head(10)
```

### Step 2: Identifikasi Masalah Data

```python
print('=== 1. MISSING VALUES ===')
missing = df_raw.isnull().sum()
persen = (missing / len(df_raw) * 100).round(2)
print(pd.DataFrame({'Jumlah Missing': missing, 'Persentase Missing (%)': persen}))


print('=== 2. DUPLICATE VALUES ===')
print(f"Jumlah baris duplicate : {df_raw.duplicated().sum()}")

print(f'\n===3. NILAI UNTUK KOLOM KATEGORIKAL ===')
for col in ['category', 'channel', 'status']:
    print(f"{col} : {df_raw[col].unique()}")

print(f"\n=== 4. QUANTITY TIDAK VALID ===")
print(df_raw[df_raw['quantity'] <= 0][['order_id', 'product_name', 'quantity']])

print(f"\n=== 5. PRODUCT NAME DENGAN SPASI ===")
spasi = df_raw[df_raw['product_name'].str.startswith('', na=False)]['product_name'].unique()
print(spasi)
```

### Step 3: Data Cleansing


| # | Masalah | Solusi |
|---|---------|--------|
| 1 | Kolom `total`, `total_sales`, `Unnamed: 15` semua kosong | Drop kolom |
| 2 | 1 baris missing values | Drop baris |
| 3 | 6 baris duplikat | Drop duplicates |
| 4 | Quantity bernilai -99 | Drop baris tidak valid |
| 5 | Spasi di `product_name` | Strip + title case |
| 6 | `category` tidak konsisten | Standarisasi title case |
| 7 | `channel` tidak konsisten | Mapping ke nama standar |
| 8 | `total` & `total_sales` kosong | Hitung ulang dari kolom lain |

```python
df = df_raw.copy()

# 1. Hapus  Kolom Tidak Relevan
df = df.drop(columns=['total', 'total_sales', 'Unnamed: 15'])
print(f"Setelah drop columns : {df.shape}")

# 2. Hapus baris dengan missing values
df = df.dropna(how='any').reset_index(drop=True)
print(f"Setelah dropna : {df.shape}")

# 3. Hapus duplikat
df = df.drop_duplicates(keep='first').reset_index(drop=True)
print(f"Setelah hapus duplikat : {df.shape}")

# 4. Hapus baris dengan quantity <= 0
df = df[df['quantity'] > 0].reset_index(drop=True)
print(f"Setelah hapus quantity tidak valid : {df.shape}")
```

```python
# --- 5. Bersihkan product_name ---
df['product_name'] = df['product_name'].str.strip().str.title()

# --- 6. Standarisasi category ---
df['category'] = df['category'].str.strip().str.title()

# --- 7. Standarisasi channel ---
channel_map = {
    'Toko_Oren_Online'   : 'Online - Toko Oren',
    'hijau_toko_web'     : 'Online - Toko Hijau',
    'Offline'            : 'Offline',
    'Online - Toko Oren' : 'Online - Toko Oren',
    'Online - Toko Hijau': 'Online - Toko Hijau'
}
df['channel'] = df['channel'].map(channel_map)

# --- Standarisasi teks lainnya ---
df['customer_name']    = df['customer_name'].str.strip().str.title()
df['customer_address'] = df['customer_address'].str.strip().str.title()

print('Nilai unik category setelah cleaning:', df['category'].unique())
print('Nilai unik channel  setelah cleaning:', df['channel'].unique())
```

### Step 4 : Data Formatting

```python
df.info()
```

```python
df = df[df['sales_date'] != '12/31/2025']
```

```python
# Format Tanggal
df['sales_date'] = pd.to_datetime(df['sales_date'], format='%d/%m/%Y')

# Ekstrak komponen tanggal
df['month'] = df['sales_date'].dt.month
df['month_name'] = df['sales_date'].dt.strftime('%B')
df['quarter'] = df['sales_date'].dt.quarter


# Perbaiki tipe data pada kolom numerik
df['price'] = df['price'].astype(float)
df['quantity'] = df['quantity'].astype(int)
df['discount'] = df['discount'].astype(float)
df['shipping_fee'] = df['shipping_fee'].astype(float)
df['rating'] = df['rating'].astype(int)


# Hitung Kolom Total dan Total Sales
df['total'] = df['price'] * df['quantity']
df['total_sales'] = df['total'] - (1 - df['discount']) + df['shipping_fee']

# Kolom Kategori Rating
df['rating_category'] = df['rating'].apply(
    lambda r: 'Positif' if r >= 4 else ('Netral' if r == 3 else 'Negatif')
)

print("=== PERBANDINGAN SEBELUM VS SESUDAH CLEANING ===")
print(f"Shape Awal : {df_raw.shape}")
print(f"Shape setelah Cleaning : {df.shape}")
print(f"Baris Dihapus : {df_raw.shape[0] - df.shape[0]}")

print()

print("=== Preview Datasets Setelah Cleaning  ==== ")
df[['order_id', 'sales_date', 'product_name', 'category', 'channel', 'quantity', 'price', 'total_sales','rating', 'rating_category', 'status']].head()

```

### Step 5 : Analisis - Jawab Business Questions

```python
# ============================================================
# Q1: Revenue & komposisi per kategori produk?
# ============================================================

rev_per_kategori = df.groupby('category').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Transaksi = ('order_id', 'count'),
    Total_Quantity = ('quantity', 'sum'),
    AVG_Transaksi = ('total_sales', 'mean')
).sort_values(by='Total_Revenue', ascending=False).reset_index()

rev_per_kategori['Share_%'] = (rev_per_kategori['Total_Revenue'] / rev_per_kategori['Total_Revenue'].sum() * 100).round(1)
rev_per_kategori['Revenue_format'] = rev_per_kategori['Total_Revenue'].apply(lambda x: f"Rp {x:,.0f}")

print("=== Q1 : REVENUE PER KATEGORI PRODUK ===")
print(rev_per_kategori[['category','Revenue_format', 'Jumlah_Transaksi', 'Total_Quantity', 'AVG_Transaksi','Share_%']])
```

```python
# ============================================================
# Q2: Channel penjualan paling menguntungkan?
# ============================================================


rev_per_channel = df.groupby('channel').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Transaksi = ('order_id', 'count'),
    AVG_Rating = ('rating', 'mean')
).sort_values('Total_Revenue', ascending=False).reset_index()

rev_per_channel['Revenue_format'] = rev_per_channel['Total_Revenue'].apply(lambda r: f"Rp {r: ,.0f}")
rev_per_channel['Share_%'] = (rev_per_channel['Total_Revenue'] / rev_per_channel['Total_Revenue'].sum() * 100).round(1)
rev_per_channel['AVG_Rating'] = rev_per_channel['AVG_Rating'].round(2)

print("=== Q2 : REVENUE PER CHANNEL ===")
print(rev_per_channel[['channel', 'Revenue_format', 'Jumlah_Transaksi', 'AVG_Rating', 'Share_%']])
```

```python
# ============================================================
# Q3: Tren penjualan per bulan?
# ============================================================


rev_per_bulan = df.groupby(['month','month_name']).agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order = ('order_id', 'count')
).reset_index().sort_values('month')

print("=== Q3 : REVENUE PER BULAN ===")
max_rev = rev_per_bulan['Total_Revenue'].max()
for _, row in rev_per_bulan.iterrows():
    bar = '█' * int(row['Total_Revenue'] / max_rev *40)
    print(f"{row['month_name']:10} | {bar: <40} | Rp {row['Total_Revenue']:< 15,.0f} ({row['Jumlah_Order']:,} orders)")
```

```python
# ============================================================
# Q4: Produk paling laris?
# ============================================================

top_produk = df.groupby('product_name').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Terjual = ('quantity', 'sum'),
    Total_Quantity = ('order_id', 'count'),
    AVG_Rating = ('rating', 'mean')
).sort_values('Total_Revenue', ascending=False).head(10).reset_index()

top_produk['Revenue_format'] = top_produk['Total_Revenue'].apply(lambda r: f"Rp {r:,.0f}")
top_produk['AVG_Rating'] = top_produk['AVG_Rating'].round(2)

print("=== Q4 : TOP 10 PRODUK PALING LARIS BERDASARKAN REVENUE ===")
print()
print(top_produk[['product_name','Revenue_format', 'Jumlah_Terjual', 'Total_Quantity',  'AVG_Rating']].to_string(index=False))
```

```python
# ============================================================
# Q5: Kota dengan penjualan tertinggi?
# ============================================================

rev_per_kota = df.groupby('customer_address').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order = ('order_id', 'count'),
    Avg_of_Order = ('total_sales', 'mean')
).sort_values('Total_Revenue', ascending=False).reset_index()

print("=== Q5 : KOTA DENGAN PENJUALAN TERTINGGI ===")
max_rev = rev_per_kota['Total_Revenue'].max()
for _ , row in rev_per_kota.iterrows():
    bar = '█' * int(row['Total_Revenue'] / max_rev *40)
    print(f"{row['customer_address']:15} | {bar: <40} | Rp {row['Total_Revenue'] : < 15,.0f} ({row['Jumlah_Order']: ,} orders) | Avg Order : Rp {row['Avg_of_Order'] :,.0f}")

print(f"\n 🏆 Kota Teratas: {rev_per_kota.iloc[0]['customer_address']} (Rp {rev_per_kota.iloc[0]['Total_Revenue']:,.0f})")
```

```python
# ============================================================
# Q6: Distribusi status order & rating pelanggan?
# ============================================================
status_dist = df['status'].value_counts().reset_index()
status_dist.columns = ['Status', 'Jumlah']
status_dist['Persentase'] = (status_dist['Jumlah'] / status_dist['Jumlah'].sum() * 100).round(1)

print('=== Q6A: DISTRIBUSI STATUS ORDER ===')
print(status_dist.to_string(index=False))

print()
rating_dist = df['rating'].value_counts().sort_index()
print('=== Q6B: DISTRIBUSI RATING PELANGGAN ===')
for rating, count in rating_dist.items():
    bintang = '⭐' * int(rating)
    bar     = '█' * int(count / rating_dist.max() * 30)
    print(f'{bintang} {bar:<30} {count:,}')

print(f'\nRata-rata rating: {df["rating"].mean():.2f} / 5')
```

```python
# ============================================================
# Q7 : Siapa Pelanggan Paling Aktif?
# ============================================================

pelanggan_aktif = df.groupby('customer_name').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order = ('order_id', 'count'),
).sort_values(['Jumlah_Order', 'Total_Revenue'], ascending=[False,False]).head(10).reset_index()

pelanggan_aktif['Total_Revenue_Format'] = pelanggan_aktif['Total_Revenue'].apply(lambda r: f"Rp {r:,.0f}")

print("=== Q7 : PELANGGAN PALING AKTIF ===")
print(pelanggan_aktif[['customer_name', 'Jumlah_Order', 'Total_Revenue_Format']].to_string(index=False))
```

```python
# ============================================================
# Q8 : Kategori Terlaris pada Setiap Channel
# ============================================================

kategori_channel = df.groupby(['channel', 'category']).agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order = ('order_id', 'count'),
    Total_Quantity = ('quantity', 'sum')
).sort_values(['channel', 'Total_Revenue'], ascending=[True,False]).reset_index()

for channel in kategori_channel['channel'].unique():
    print(f"\nChannel : {channel}")

    subset = kategori_channel[kategori_channel['channel'] == channel]

    for _, row in subset.iterrows():
        print(f"{row['category']:<15} | Rp {row['Total_Revenue']:>15,.0f}   ({row['Jumlah_Order']:,} Orders) | Total Qty : {row['Total_Quantity']:,}")
```
